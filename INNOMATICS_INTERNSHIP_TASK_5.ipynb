Problem 1: Social Media – Post Engagement Analyze

```python
def analyze_engagement(posts):
    post_no = 1
    for likes in posts:
        total_likes = sum(likes)

        if total_likes >= 1000:
            status = "Viral Post"
        else:
            status = "Normal Engagement"
        print("Post", post_no)
        print("Total Likes:", total_likes)
        print("Post Status:", status)
        print("Next post status is")
        post_no += 1
analyze_engagement([
    [200, 300, 150, 400],   # Post 1
    [100, 150, 200],        # Post 2
    [500, 300, 250]         # Post 3
])
```

Problem 3:Agriculture – Rainfall Adequacy Checker


```python
def rainfall_checker(rainfall, required):
    avg = sum(rainfall) / len(rainfall)
    if avg >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print("Average Rainfall:", int(avg))
    print("Rainfall Status:", status)
rainfall_checker([70, 75, 80, 63], 70)
```

Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
detect_duplicates(["john", "alex", "john", "emma"])
```

Problem 5: Healthcare – Appointment Eligibility Checker

```python
def check_eligibility():
  n = int(input("Enter number of patents: "))
  for i in range(1,n+1):
    age = int(input("Enter patient age {i}: "))
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    print("Patient Age:", age)
    print("Eligibility Status:", status)
check_eligibility()
```

Problem 7: System Monitoring – Application Health Checker

```python
def health_error():
    n = int(input("Enter number of errors to check: "))
    for i in range(1, n+1):
        errors = int(input(f"Enter error count {i}: "))
        if errors == 0:
            status = "Healthy"
        elif errors <= 5:
            status = "Minor Issues"
        else:
            status = "Critical Issues"
        print("Error Count:", errors)
        print("System Status:", status)
        print("-------------")
health_error()
```

Problem 8: Banking – Daily Transaction Limit Checker

```python
def transaction_check():
  n = int(input("Enter number of transactions: "))
  for i in range(1, n+1):
    amount = int(input(f"Enter transaction amount {i}: "))
    limit = 50000
    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)
transaction_check()
```

Problem 9: E-Learning – Student Attendance Eligibility System

```python
def attendance_check():
    n = int(input("Enter days: "))
    present = 0
    for i in range(1, n+1):
        s = input(f"Day {i} (P/A): ").strip().lower()
        if s == "p" or s == "present":
            present += 1
    percentage = (present / n) * 100
    if percentage >= 75:
        print("Attendance Percentage:", percentage)
        print("Exam Eligibility: Eligible")
    else:
        print("Attendance Percentage:", percentage)
        print("Exam Eligibility: Not Eligible")
attendance_check()
```

Problem 10: Smart Electricity Bill Analyzer


```python
def electricity_bill():
  units = int(input("Enter electricity units: "))
  bill = 0
  if units <= 100:
        bill = units * 3
  elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
  else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
  if bill < 500:
        usage = "Low Usage"
  elif bill <= 1500:
        usage = "Moderate Usage"
  else:
        usage = "High Usage"
  return bill, usage
amount, status = electricity_bill()
print("Total Bill:", amount)
print("Usage Level:", status)
```
