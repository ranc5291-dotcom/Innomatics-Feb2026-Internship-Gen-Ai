Problem 1: Smart Parking Lot Management System

```python
def smart_parking(capacity, logs):
    parked = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1
    print("Currently Parked Vehicles:", parked)
    if parked > capacity:
        print("Parking Status: Full (Alert!)")
    else:
        print("Parking Status: Available")
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking(capacity, logs)
```

Problem 2: Online Food Delivery Time Estimator

```python
def estimate_delivery(distance, traffic, weather):
    time = distance * 5
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10
    if weather == "Rainy":
        time += 10
    elif weather == "Storm":
        time += 20
    print("Estimated Delivery Time:", time, "minutes")
distance = float(input("Enter distance (km): "))
traffic = input("Enter traffic level (Low/Medium/High): ")
weather = input("Enter weather (Clear/Rainy/Storm): ")
estimate_delivery(distance, traffic, weather)
```

Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy(total_seats, booked_list):
    booked = sum(booked_list)
    occupancy = (booked / total_seats) * 100
    print("Occupancy:", int(occupancy), "%")
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")
booked_seats = [1] * 150
seat_occupancy(200, booked_seats)
```

Problem 4: Cloud Server Load Classification System

```python
def server_load(cpu_readings):
    avg = sum(cpu_readings) / len(cpu_readings)
    print("Average CPU Load:", int(avg), "%")
    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
server_load([45, 60, 70, 85, 90])
```

Problem 5: Smart Classroom Resource Usage Monitor

```python
def resource_monitor(usage):
    overused = []
    for resource, hours in usage.items():
        if hours > 8:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
resource_monitor(usage_data)
```

Problem 6: Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):
    if registrations <= capacity:
        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")
    else:
        print("Confirmed Registrations:", capacity)
        print("Waitlisted Users:", registrations - capacity)
        print("Registration Status: Closed")
event_registration(100, 105)
```
