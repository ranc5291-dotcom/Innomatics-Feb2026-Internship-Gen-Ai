2.PASS / FAIL ANALYZER

```python
marks = [45, 78, 90, 33, 60] # List of student marks
pass_count = 0  # Variable to count passed students
fail_count = 0  # Variable to count failed students
for m in marks:
    if m >= 50:
        pass_count += 1 # Increase pass count by 1
    else:
        fail_count += 1 # Increase fail count by 1
print("Total Passed Students:", pass_count) # Print results
print("Total Failed Students:", fail_count) # Print results

```

3. SIMPLE DATA CLEANER

```python
names = [" Alice ", "bob", " CHARLIE "]
print([n.strip().lower() for n in names]) # 1. strip()  -> removes leading and trailing spaces
                                          # 2. lower()  -> converts the name to lowercase

```

4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:  # Loop through each message
    length = len(msg) # Find the length of the message
    print(f"Message: '{msg}' | Length: {length}")  # Print the message and its length
if length > 10:          # Check if the message is longer than 10 characters
        print("Flag: Message is longer than 10 characters")
```
