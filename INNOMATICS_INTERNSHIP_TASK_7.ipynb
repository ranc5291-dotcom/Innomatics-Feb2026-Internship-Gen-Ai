Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"
    def set_status(self, action):
        self.status = action
    def display_status(self):
        print(f"{self.name} is {self.status}")
light_name = input("Light Name: ")
action = input("Action: ")
light = SmartLight(light_name)
light.set_status(action)
light.display_status()
```

Problem 2: Employee ID Card System

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")
emp = Employee(name, emp_id, department)
emp.display_id_card()
```

Problem 3: Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)
name = input("Contact Name: ")
phone = input("Phone Number: ")
c = Contact(name, phone)
c.display_contact()
```

Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))
name = input("Product Name: ")
price = input("Price: ")
p = Product(name, price)
p.display_price_tag()
```

Problem 5: Movie Rating Display System

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")
name = input("Movie Name: ")
rating = input("Rating: ")
m = Movie(name, rating)
m.display_movie()
```

Problem 6: Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address
    def display_details(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)
customer = input("Customer Name: ")
address = input("Address: ")
d = Delivery(customer, address)
d.display_details()
```
