Problem 1: Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    while True:
        amount = int(input("Enter recharge amount: "))
        if amount < 50:
            print("❌ Recharge must be at least ₹50. Try again.")
        elif amount not in valid_plans:
            print("❌ Invalid plan. Valid plans are:", valid_plans)
        else:
            print("✅ Recharge successful for ₹", amount)
            break
validate_recharge()
```

Problem 2: Inventory Reorder Alert System

```python
def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")
inventory = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 30
}
check_inventory(inventory)
```

Problem 3: Student Result Processing System

```python
def student_result(marks):
    total = 0
    for m in marks:
        total += m

    avg = total / len(marks)

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"

student1 = [60,45,55,50]
student2 = [45,40,45,70]
student3 = [20,10,25,30]
print("Student 1 Result:", student_result(student1))
print("Student 2 Result:", student_result(student2))
print("Student 3 Result:", student_result(student3))
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (distance * 12)
    if peak == "yes":
        fare = fare + (fare * 0.25)
    return fare
while True:
    km = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()
    total_fare = calculate_fare(km, peak)
    print("Total Fare: ₹", total_fare)
    retry = input("Do you want to retry? (yes/no): ").lower()
    if retry != "yes":
        break
```

Problem 5: Employee Attendance Eligibility Checker

```python
def attendance_checker(attendance):
    present_count = 0
    for day in attendance:
        if day == "P":
            present_count += 1
    percentage = (present_count / len(attendance)) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance_data = {
    "Ravi": ["P", "P", "A", "P", "P", "A", "P"],
    "Anita": ["P", "A", "A", "P", "A", "P", "A"],
    "Kiran": ["P", "P", "P", "P", "P", "A", "P"]
}

for name, record in attendance_data.items():
    result = attendance_checker(record)
    print(name, ":", result)
```

Problem 6: Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$*&!"
    if len(password) < 8:
        return "Weak Password"
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
while True:
    pwd = input("Enter password: ")
    result = check_password(pwd)
    print(result)
    if result == "Strong Password":
        break
    else:
        print("❌ Weak password. Please try again.\n")
```
