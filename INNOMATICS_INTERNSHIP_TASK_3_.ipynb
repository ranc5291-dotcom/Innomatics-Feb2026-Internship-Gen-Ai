Problem 1: Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
max_score = max(employees.values())
top_performers = [name for name, score in employees.items() if score == max_score]
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score is: {max_score})")
```

Problem 2: Search Query Keyword Analysis

```python
query = "Buy mobile phone buy phone online"
words = query.lower().split()
count = {}
for w in words:
    count[w] = count.get(w, 0) + 1
result = {w: c for w, c in count.items() if c > 1}
print(result)
```

Problem 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
count = {}
for e in emails:
    domain = e.split("@")[1]
    count[domain] = count.get(domain, 0) + 1
total = len(emails)
for d in count:
    print(d, ":", int((count[d] / total) * 100), "%")
```

Problem 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg = sum(sales) / len(sales)
for i in range(len(sales)):
    if sales[i] > avg * 1.3:   # 30% above average
        print("Day", i+1, ":", sales[i])
```

Problem 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
id_count = {}
for uid in user_ids:
    id_count[uid] = id_count.get(uid, 0) + 1
for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
